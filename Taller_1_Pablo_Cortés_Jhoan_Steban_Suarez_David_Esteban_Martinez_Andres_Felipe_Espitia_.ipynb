##1) Propósito
Permitir que un usuario juegue “piedra-papel-tijeras” contra la máquina. La máquina elige al azar; el usuario ingresa su elección. Se indica quién gana cada ronda y el primer jugador en llegar a 3 victorias (mejor de 5) gana el juego.
##2) Entradas
Del usuario (por ronda): Una opción entre (piedra, papel, tij
eras) (variantes : p , pa, t)

Del sistema:la máquina genera su elección aleatoria
##3) Salidas
Mensaje con las elecciones de usuario y máquina por ronda.
Resultado de cada ronda: punto para usuario / punto para maquina / empate
Marcador acumulado (ej: Usuario 2 - 1 Máquina)
Resultado final de la serie : ganador del juego.
##4) Reglas del juego
piedra gana a tijeras
papel gana a piedra
tijeras vence a papel
misma elección ⇒ empate
##5) Restricciones
Implementación modular (basada en funciones).
Debe poder ejecutarse en Jupyter Notebook y llevar nombres de integrantes en el notebook y en el PDF.
Comentarios en el código.

##CODIGO FUENTE

```python
import random

# Función que genera la elección de la máquina de manera aleatoria
def eleccion_maquina():
    # La máquina elige al azar entre piedra, papel o tijera
    opciones = ["piedra", "papel", "tijera"]
    return random.choice(opciones)

# Función que determina el ganador de una ronda
def ganador_ronda(usuario, maquina):
    # Si ambos eligen lo mismo, es empate
    if usuario == maquina:
        return "empate"
    # Condiciones en las que gana el usuario
    elif (usuario == "piedra" and maquina == "tijera") or \
         (usuario == "tijera" and maquina == "papel") or \
         (usuario == "papel" and maquina == "piedra"):
        return "usuario"
    # Si no es empate y no gana el usuario, gana la máquina
    else:
        return "maquina"

# Función principal que maneja el juego completo (mejor de 5 rondas, a 3 puntos)
def jugar():
    # Contadores de puntos para usuario y máquina
    puntos_usuario = 0
    puntos_maquina = 0

    # El juego continúa hasta que alguno llegue a 3 puntos
    while puntos_usuario < 3 and puntos_maquina < 3:
        # Se pide al usuario su elección y se convierte a minúsculas
        usuario = input("Elige piedra, papel o tijera: ").lower()
        # La máquina realiza su elección
        maquina = eleccion_maquina()
        print("La máquina eligió:", maquina)

        # Se determina el ganador de la ronda
        ganador = ganador_ronda(usuario, maquina)

        # Se actualizan los puntajes según el resultado
        if ganador == "usuario":
            puntos_usuario += 1
            print("Ganaste esta ronda")
        elif ganador == "maquina":
            puntos_maquina += 1
            print("Perdiste esta ronda")
        else:
            print("Empate")

        # Mostrar el marcador después de cada ronda
        print("Marcador -> Usuario:", puntos_usuario, "Maquina:", puntos_maquina, "\n")

    # Cuando alguien llega a 3 puntos, se determina el ganador del juego
    if puntos_usuario == 3:
        print("¡Ganaste el juego!")
    else:
        print("La máquina ganó el juego")

# Llamada a la función principal para iniciar el juego
jugar()
```
